### Promedio y máximo de solicitudes por unidad de tiempo (1min)

```python
#Pasa a un formato de fecha correcto
df["request_time"] = pd.to_datetime(df['request_time'], format='%Y-%m-%d %H:%M:%S.%f %z')
```

```python
# Esta función genera un df en el que cada fila contiene el total de solicitudes en un minuto, desde que se hizo la primera.
df_minuto = df.groupby(pd.Grouper(key='request_time', freq='1Min')).size().reset_index(name='solicitudes')
```

### Promedio y máximo de UAS volando simultaneamente en el mismo espacio aéreo.

```python
#Se cogen estas dos columnas del df y se asigna cada una a una variable
columna1 = df['takeoff_time_EST'] 
columna2 = df['landing_time_EST'] 

# Crear una columna que indica el origen de cada valor
origen = ['despegue'] * len(columna1) + ['aterrizaje'] * len(columna2)

# Concatenar las dos columnas una encima de la otra
nueva_columna = pd.concat([columna1, columna2], ignore_index=True)

# Crear un nuevo df con las columnas 'valor' y 'origen'
nuevo_df = pd.DataFrame({'valor': nueva_columna, 'origen': origen})

print(nuevo_df)
```

```python
#Ordenar el df por tiempo en el que se han realizado las operaciones.
nuevo_df.sort_values(by='valor')
```

```python
uas_volando = 0
max_uas_volando = 0
av = 0
c=0

# Iterar a través de cada fila
for _, row in nuevo_df.iterrows():
    if row['origen'] == 'despegue':
        #En el momento que despega un dron se suma un token a la var uas_volando
        uas_volando += 1
    else:
        #En el momento que aterriza un dron se resta un token a la var uas_volando
        uas_volando -= 1
    # Actualiza el máximo recuento de uas volando simultáneamente
    max_uas_volando = max(max_uas_volando, uas_volando)
    av = av+uas_volando
    c+=1 #Total de operaciones 

av = av/c 
print("Máximo número de UAS volando simultáneamente: ", max_uas_volando)
print("El promedio de UAS volando simultáneamente: ", av)
```

### Dibujar la ruta 


```python
#Api par introduir el fons de google maps.
import os
import types

import cartopy.io.img_tiles as img_tiles
import requests
import PIL

class CachedTiler(object):
    def __init__(self, tiler):
        self.tiler = tiler

    def __getattr__(self, name):
        attr = getattr(self.tiler, name, None)
        if isinstance(attr, types.MethodType):
            attr = types.MethodType(attr.__func__, self)
        return attr

    def get_image(self, tile):
        tileset_name = '{}'.format(self.tiler.__class__.__name__.lower())
        cache_dir = os.path.expanduser(os.path.join('~/', 'image_tiles', tileset_name))
        if not os.path.exists(cache_dir):
            os.makedirs(cache_dir)
        tile_fname = os.path.join(cache_dir, '_'.join(str(v) for v in tile) + '.png')
        if not os.path.exists(tile_fname):
            response = requests.get(self._image_url(tile),
                                    stream=True)

            with open(tile_fname, "wb") as fh:
                for chunk in response:
                    fh.write(chunk)
        with open(tile_fname, 'rb') as fh:
            img = PIL.Image.open(fh)
            img = img.convert(self.desired_tile_form)     
        return img, self.tileextent(tile), 'lower'
```

```python
uas = input("Introdueix uuid del UAS: ")
```

```python
# Cadena de entrada con las coordenadas de los drones
fila = df[df["uuid"] == uas]
cadena_coordenadas = fila["geometry"].iloc[0]

# Eliminar la parte inicial y final de la cadena para obtener solo las coordenadas
cadena_coordenadas = cadena_coordenadas.replace("POLYGON Z((", "").replace("))", "")

# Separar las coordenadas por comas y luego dividir cada coordenada en latitud y longitud
coordenadas = cadena_coordenadas.split(", ")
coordenadas = [coord.split()[:2] for coord in coordenadas]

# Convertir las cadenas a flotantes y crear la lista de coordenadas
coordenadas_drones = [(float(coord[0]), float(coord[1])) for coord in coordenadas]

print(coordenadas_drones)
```

```python
import gmplot
for i in coordenadas_drones:
    # Centra el mapa en el primer par de coordenadas
    gmap = gmplot.GoogleMapPlotter(i[0], i[1], zoom=6)

    # Dibuja la ruta del dron
    latitudes, longitudes = zip(*coordenadas_drones)
    gmap.plot(latitudes, longitudes, 'cornflowerblue', edge_width=10)

# Guarda el mapa en un archivo HTML
gmap.draw("ruta_dron.html")
```
